## Genera 
The two genera that have appeared in my work are L and $\hat A$.
Recall from Milnor-Stasheff $\S19$ or Hirzebruch the notion of a multiplicative sequence $K_n$. These are homogenous polynomials wuch that when summed they form a graded ring map (more or less). There is a bijection between these multiplicative sequences and formal power series, given by evaluating the multiplicative sequence (the sum of the terms) on $1+t$. Any such multiplicative sequence defines a genus, that is a map from manifolds to (say) the rationals which is zero for manifolds that are not $4k$ dimensional and otherwise $K(M^{4k}) = K_n(p_1, ..., p_k)[M]$ evaluating the polynomial on the Pontryagin classes of the manifold (of the tangent bundle) and then on the fundamental class of the manifold. Any such genera will descend to the cobordism ring. Because the Pontryagin numbers are complete invariants on the _rational_ cobordism ring it can be shown that any ring homomorphism from the rational cobordism ring corresponds to a genus.

The signature is such a ring map and therefore it has a genus that is called the L-genus. Its power series is given by the Taylor series of
$$\frac{\sqrt{t}}{\tanh \sqrt{t}}$$
 We want a program that will print the L polynomials.

Another genus is the A hat genus, which is given by the power series
$$\frac{\frac{1}{2}\sqrt{t}}{\sinh \frac{1}{2} \sqrt{t}}$$

The only other genus I have come accross is the Todd genus in Hirzebruchs book.

## Characters
In opposition to genera we also have additive sequences which can form so called characters. The Chern character is (basically the only) the example of this, its power series is given by $e^t$.

## Defining a useful function

```python
def genus(n, f, t):
    '''Return the K_n(p_1, ..., p_n) polynomial (multiplicative series) for a function f, in variable t'''
    R = PolynomialRing(QQ, names=[f'p{i}' for i in range(1, n + 1)])
    K_n = R.zero()
    
    Sym = SymmetricFunctions(QQ)
    m = Sym.monomial()
    e = Sym.e() # Initialize the elementary symmetric function basis
    
    # Use .coefficient() for safety, as .list() can truncate trailing zeros
    f_coefficients = [f.series(t, 2*n + 2).coefficient(t, 2*k) for k in range(n + 1)]

    for q in Partitions(n):
        lambda_I = prod(f_coefficients[i] for i in q)
        
        # 1. Convert the monomial symmetric function to the elementary basis
        m_in_e = e(m[q])
        
        # 2. Map e_k to the Pontryagin class p_k
        # m_in_e iterates as pairs of (partition mu, coefficient c)
        # A partition mu like [2, 1] represents the term e_2 * e_1
        q_poly = sum(c * prod(R.gen(k - 1) for k in mu) for mu, c in m_in_e)
        
        K_n += lambda_I * q_poly

    return K_n
```

```python
def character(n, f, t):
    '''Return the additive sequence corresponding to a power series'''
    if n == 0:
        print("The 0-th term evaluates to the rank of the bundle")
        return 0 # The 0-th term evaluates to the rank of the bundle, handled outside this polynomial
        
    R = PolynomialRing(QQ, names=[f'p{i}' for i in range(1, n + 1)])
    
    Sym = SymmetricFunctions(QQ)
    p = Sym.p() # Initialize the power sum symmetric function basis
    e = Sym.e() # Initialize the elementary symmetric function basis
    
    # Extract the 2n-th coefficient of the power series
    a_n = f.series(t, 2*n + 1).coefficient(t, 2*n)

    # 1. Express the n-th power sum symmetric function in the elementary basis
    s_in_e = e(p[n])
    
    # 2. Map e_k to the Pontryagin class p_k
    s_poly = sum(c * prod(R.gen(k - 1) for k in mu) for mu, c in s_in_e)
    
    return a_n * s_poly
```

# Stolz
First we define the invariant, then we evaluate it on plumbings whose Pontrygin classes are known by Bott.

```python
from sympy import bernoulli
from sympy import factorint
from math import floor
```

```python
def a(n):
    if n%2 == 0:
        return 1
    else:
        return 2

def epsilon(k):
    return (3 - (-1)**k)//2

def mersene(k):
    return int(2**(2*k - 2) * (2**(2*k - 1) - 1))

def bP_4(k):
    bern = (-1)**(k+1)*bernoulli(2*k) # Topologists Bernoulli numbers
    bern = bern/(4*k)
    num_bern = bern.denominator()
    return epsilon(k)*mersene(k)*num_bern

def bezout(x,y):
    g, a, b = xgcd(x,y)
    return a , b

def extract_graded_part(poly, target_degree):
    R = poly.parent()
    result = R.zero()

    # Assuming you have a ring R and a symbolic expression 'expr'
    R = PolynomialRing(QQ, names=[f'p{i}' for i in range(1, n + 1)])
    
    # Coerce the symbolic expression into a strict polynomial
    my_poly = R(poly)
    
    # poly.dict() returns {(exp_p1, exp_p2, ...): coefficient}
    for exps, coeff in my_poly.dict().items():
        # Calculate graded degree: sum of (index * exponent)
        # enumerate(..., start=1) pairs index k with exponent e
        term_degree = sum(k * e for k, e in enumerate(exps, start=1))
        
        if term_degree == target_degree:
            # Reconstruct the monomial and add it to the result
            monomial = prod(R.gen(k-1)**e for k, e in enumerate(exps, start=1))
            result += coeff * monomial
            
    return result


def A_times_ph(n):
    t = var('t')
    A_power = ((1/2)*t) / sinh((1/2)*t)
    ph_power = 2*cosh(t)
    
    R = PolynomialRing(QQ, names=[f'p{i}' for i in range(1, n + 1)])
    total_A = R.zero() + 1 #p_0 is 1
    total_ph = R.zero() + n #ph_0 is the rank of the bundle

    #Get the total A and ph
    for i in range(1, n):
        total_A += genus(i, A_power, t)
        total_ph += character(i, ph_power, t)

    #multiply the total classes
    total = total_A * total_ph

    #find only the graded degree n pieces
    return extract_graded_part(total, n)
    
```

```python
def stolz(n):
    #initialise all the power series that appear in the invariant
    t = var('t')
    L_power = t/tanh(t)
    A_power = ((1/2)*t) / sinh((1/2)*t)
    ph_power = 2*cosh(t)

    #initialise the ring and polynomial
    #R = PolynomialRing(QQ, names=[f'p{i}' for i in range(1, n + 1)])
    #Im worried about the name spaces of the ring and the interaction inside and outside the differnt calls of the function
    bern = (-1)**(n+1)*bernoulli(2*n) # Topologists Bernoulli numbers
    bern = bern/(4*n)
    num_bern = bern.denominator()
    den_bern = bern.numerator()
    c,d = bezout(num_bern, den_bern)
        
    return (1/8)*genus(n, L_power, t) + (bP_4(n)/a(n))*(c*genus(n, A_power, t) + ((-1)**n)*d*A_times_ph(n))
```

Now we need to evaluate the polynomial on the known Pontryagin classes of the plumbings before we can reduce modulo $bP_{4k}$

```python
def pont_plumb(n,m,k):
    #Return the Pontryagin number / class p_k of the sphere bundle that corresponds
    #to the homotopy class in \pi_m SO_n \times \pi_n SO_m. Note that m = 4r, n=4k - 4r, 
    #otherwise the Pontryagin class will be zero. Stolz formula doesnt contain the top 
    #Pontryagin class so we dont need to worry about the behaviour there. '''
    if m%4 == 0 and n+m%4 ==0:
        if k==m//4:
            return a(m//4)*((2*(m//4) - 1).factorial()) 
        elif k==n//4:
            return a(n//4)*((2*(n//4) - 1).factorial()) 
    else:
        return 0

def total_pont_plum(n,m):
    nums = []
    for i in range(1,n+m+2):
        nums.append(pont_plumb(n,m,i))
    return nums
```

```python
from sage.all import *

def a(r):
    '''Coefficient for Bott periodicity generators.'''
    return 1 if r % 2 == 0 else 2

def extract_graded_part(poly, target_degree):
    # Define the ring first using target_degree instead of n
    R = PolynomialRing(QQ, names=[f'p{i}' for i in range(1, target_degree + 1)])
    
    # Coerce the input into the strict polynomial ring
    my_poly = R(poly)
    
    # Now R is defined and can be used to create the zero polynomial
    result = R.zero()
    
    for exps, coeff in my_poly.dict().items():
        term_degree = sum(k * e for k, e in enumerate(exps, start=1))
        
        if term_degree == target_degree:
            monomial = prod(R.gen(k-1)**e for k, e in enumerate(exps, start=1))
            result += coeff * monomial
            
    return result

from sage.all import bernoulli, xgcd

def get_stolz_coefficients(n):
    '''Calculates the exact constants required for the Stolz S_n function.'''
    a_n = 1 if n % 2 == 0 else 2
    
    # Topologists' Bernoulli numbers (B_1 = 1/6, B_2 = 1/30, etc.)
    B_n = abs(bernoulli(2 * n))
    frac = B_n / (4 * n)
    num = frac.numerator()
    denom = frac.denominator()
    
    # Extended Euclidean Algorithm to find c_n and d_n such that: c_n*num + d_n*denom = 1
    # xgcd returns (gcd, c_n, d_n)
    _, c_n, d_n = xgcd(num, denom)
    
    # Calculate the order of bP_4n
    bP_order = a_n * (2**(2*n - 2)) * ((2**(2*n - 1)) - 1) * num
    
    return a_n, bP_order, c_n, d_n

def evaluate_milnor_plumbing(k, r):
    '''
    Evaluates the Stolz invariant for the plumbing of an (4r-1)-sphere 
    bundle over a (4(k-r)-1)-sphere.
    '''
    # 1. Determine the generators for the Pontryagin homomorphism
    p_r_f1 = a(r) * factorial(2*r - 1)
    p_kr_f2 = a(k - r) * factorial(2*(k - r) - 1)
    
    # 2. Explicitly define the polynomial ring
    R = PolynomialRing(QQ, names=[f'p{i}' for i in range(1, k + 1)])
    
    # 3. Get the Stolz polynomial and force it into the strict polynomial ring
    S_poly = R(stolz(k))
    
    # 4. Find the specific monomial p_r * p_{k-r}
    target_exps = [0] * R.ngens()
    if r == k - r:
        target_exps[r - 1] = 2
        W_eval = 2 * p_r_f1 * p_kr_f2 
    else:
        target_exps[r - 1] = 1
        target_exps[k - r - 1] = 1
        W_eval = p_r_f1 * p_kr_f2
        
    # Extract the coefficient for this specific monomial
    coeff = S_poly.dict().get(tuple(target_exps), 0)
    
    # 5. Evaluate the polynomial on W (all other terms evaluate to 0)
    S_n_value = coeff * W_eval
    
    # 6. Calculate the final invariant
    _, bP_order, _, _ = get_stolz_coefficients(k)
    s_sigma = (1/8) * 0 - S_n_value
    
    return mod(s_sigma, bP_order)

# To find non-standard structures on S^19 (k=5), using r=2:
# print(evaluate_milnor_plumbing(5, 2))
```
